# Alzheimer's EEG Dataset Statistics
This notebook demonstrates how to load the Kaggle dataset `codingyodha/largest-alzheimer-eeg-dataset` and extract its key statistics.

```python
import numpy as np
import kagglehub

# Download the dataset
path = kagglehub.dataset_download("codingyodha/largest-alzheimer-eeg-dataset")
print("Dataset downloaded to:", path)
```

```python
import os
dataset_path = os.path.join(path, "integrated_eeg_dataset.npz")

print("Loading dataset...")
data = np.load(dataset_path, allow_pickle=True)
X = data['X_raw']
y = data['y_labels']

print(f"Data shape X_raw: {X.shape}")
print(f"Labels shape y_labels: {y.shape}")
```

## Calculating Statistics

```python
# Map string labels to numeric for checking distributions
labels = np.array([float(x) for x in y[:, 0]])

# Create global unique ID by combining dataset source + subject ID
global_sub_ids = [f"{src}_{sub}" for src, sub in zip(y[:, 2], y[:, 1])]
unique_subjects = np.unique(global_sub_ids)

total_subjects = len(unique_subjects)

hc_count = 0
ad_count = 0

for L in np.unique(labels):
    mask = labels == L
    subs = np.unique(np.array(global_sub_ids)[mask])
    if L == 0.0:
        hc_count = len(subs)
    elif L == 1.0:
        ad_count = len(subs)

duration_seconds = len(y) # Each epoch is 1 second
sampling_rate = X.shape[1]
channels = X.shape[2]

print("=== Dataset Statistics ===")
print(f"Total number of subjects: {total_subjects}")
print(f"Number of Healthy Controls (HC): {hc_count}")
print(f"Number of Alzheimer's Disease (AD) subjects: {ad_count}")
print(f"Number of EEG channels: {channels}")
print(f"Sampling rate (Hz): {sampling_rate}")
print(f"Recording duration: {duration_seconds} seconds (~{duration_seconds/3600:.1f} hours)")
print(f"Shape of data matrix per epoch: {channels} Channels x {sampling_rate} Time samples")
```
